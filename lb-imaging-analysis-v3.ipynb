## ML Classification

```python
# ML Pipeline Function
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score

# Logistic Regression (L1 and L2) and Lasso
from sklearn.linear_model import LogisticRegression, Lasso
# Random Forest 
from sklearn.ensemble import RandomForestClassifier
# Support Vector Machine
from sklearn.svm import SVC
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB, MultinomialNB
# Artificial Neural Network (Multi-Layer Perceptron)
from sklearn.neural_network import MLPClassifier

def pipeline_predict(X_train, X_test, y_train, y_test):
    pipelines = {
    'l1-lr':make_pipeline(StandardScaler(), LogisticRegression(penalty='l1',solver='liblinear',max_iter=250, random_state=0)), # solver='saga' or 'liblinear'
    'l2-lr':make_pipeline(StandardScaler(), LogisticRegression(multi_class='multinomial',penalty='l2',solver='lbfgs',max_iter=300, random_state=0)),
    #'lasso':make_pipeline(StandardScaler(), Lasso()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier(random_state=0)),
    'svm':make_pipeline(StandardScaler(), SVC(random_state=0)),
    'gnb':make_pipeline(StandardScaler(), GaussianNB()),
    'mnb':make_pipeline(MinMaxScaler(), MultinomialNB()),
    'ann':make_pipeline(StandardScaler(), MLPClassifier(max_iter=1400, random_state=0)),    
    }
    
    fit_models = {}
    for algo, pipeline in pipelines.items():
        model = pipeline.fit(X_train, y_train)
        fit_models[algo] = model
    
    algorithms = []
    results = []
    for algo, model in fit_models.items():
        yhat = model.predict(X_test)
        #print(algo, accuracy_score(comb_y_test, yhat))
        algorithms.append(algo)
        results.append(accuracy_score(y_test, yhat))
    return algorithms, results
```

```python
# Split the dataset into features and labels
# MRI Features to use per discussion with Mateusz
mri_X = _mri[['Total Tumor Volume (mm^3)',
     'Enhancing (mm^3)', 'Non-enhancing (mm^3)',
     'Cystic (core) (mm^3)', 'Cystic (reactive) (mm^3)', 
     'Edema (mm^3)','Adjacent surface (y/n)',
     'Leptomeningeal disease (y/n)','Tumor location (general)']].values

mri_y = _mri['Short_histology'].values.ravel()

# Split data into training and testing set
mri_X_train, mri_X_test, mri_y_train, mri_y_test = train_test_split(mri_X, mri_y, test_size=0.25, random_state=0)

#Sanity check
print(mri_X_train.shape, mri_X_test.shape, mri_y_train.shape, mri_y_test.shape)
```

```python
pipeline_predict(mri_X_train, mri_X_test, mri_y_train, mri_y_test)
```
